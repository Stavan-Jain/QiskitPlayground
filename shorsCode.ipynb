```python
# Building up to Shor's Code
from qiskit import QuantumCircuit, QuantumRegister
from qiskit import ClassicalRegister
from qiskit.quantum_info import SparsePauliOp
from qiskit.quantum_info import Statevector
```

# Correcting Bit-Flip Errors

```python
# bit 0 : data qubit
# bit 1, 2 : used for 3-bit repetition code
# bit 3 : ancilla qubit used to diagnose syndrome

data = QuantumRegister(3, "data")
anc  = QuantumRegister(1, "anc")
qc = QuantumCircuit(data, anc)
# preparing the |+> state for the data qubit
qc.h(0)

# duplicate the data qubit
qc.cx(0, 1)
qc.cx(0, 2)

# syndrome detection
qc.cx(0, 3)
qc.x(1)
qc.cx(1, 3)
qc.add_register(ClassicalRegister(2, "syndrome"))
sv = Statevector.from_instruction(qc)

qc.measure(3, qc.clbits[0])
qc.reset(3)
qc.cx(1, 3)
qc.cx(2, 3)
qc.measure(3, qc.clbits[1])
qc.draw("mpl")
```

# Correcting Phase-Flip Errors

```python
data = QuantumRegister(3, "data")
anc  = QuantumRegister(1, "anc")
qc = QuantumCircuit(data, anc)
# preparing the |+> state for the data qubit
qc.h(0)

# duplicate the data qubit
qc.cx(0, 1)
qc.cx(0, 2)

# express the data in the Hadamard basis
qc.h(0)
qc.h(1)
qc.h(2)

# syndrome detection
qc.cx(0, 3)
# qc.x(1)
qc.cx(1, 3)
qc.add_register(ClassicalRegister(2, "syndrome"))
sv = Statevector.from_instruction(qc)

qc.measure(3, qc.clbits[0])
qc.reset(3)
qc.cx(1, 3)
qc.cx(2, 3)
qc.measure(3, qc.clbits[1])

# rotate the data back, to express it in the computational basis
qc.h(0)
qc.h(1)
qc.h(2)
qc.draw("mpl")
```

# Shor's Code

```python
block1 = QuantumRegister(3, "block_a")
block2  = QuantumRegister(3, "block_b")
block3  = QuantumRegister(3, "block_c")
qc = QuantumCircuit(block1, block2, block3)
# preparing the |+> state for the data qubit
#qc.h(0)

# duplicate the data qubit for bit-flip detection
qc.cx(0, 3)
qc.cx(0, 6)

# express the data in the Hadamard basis
qc.h(0)
qc.h(3)
qc.h(6)

# duplicate the data qubits for phase-flip detection
def duplicate(index):
    qc.cx(index, index + 1)
    qc.cx(index, index + 2)

duplicate(0)
duplicate(3)
duplicate(6)

qc.draw("mpl")
```
